### Single Action Potential Data Preprocessing

```python

def PackageData_SingleAP(coupling_type, filepath, folder_to_write):

    """Function created to package data from single AP sinulation for all coupling types.
        Data is written to "data/singleAP/ directory.
    """
    
    rel_vesicles_WT = pd.DataFrame()
    rel_rate_WT = pd.DataFrame()
    reserve_pool_rate_WT = pd.DataFrame()
    docked_pool_rate_WT = pd.DataFrame()
    aysnc_rel_rate_WT = pd.DataFrame()
    sync_rel_rate_WT = pd.DataFrame()
    spont_rel_rate_WT = pd.DataFrame()
    slow_rel_rate_WT = pd.DataFrame()
    fast_rel_rate_WT = pd.DataFrame()
    rel_proba_WT = pd.DataFrame()
    RRV_WT = pd.DataFrame()
    Ca_VGCC_WT = pd.DataFrame()
    Ca_CYTO_WT = pd.DataFrame()
    Ca_IP3_WT = pd.DataFrame()

    rel_vesicles_AD = pd.DataFrame()
    rel_rate_AD = pd.DataFrame()
    reserve_pool_rate_AD = pd.DataFrame()
    docked_pool_rate_AD = pd.DataFrame()
    aysnc_rel_rate_AD = pd.DataFrame()
    sync_rel_rate_AD = pd.DataFrame()
    spont_rel_rate_AD = pd.DataFrame()
    slow_rel_rate_AD = pd.DataFrame()
    fast_rel_rate_AD = pd.DataFrame()
    rel_proba_AD = pd.DataFrame()
    RRV_AD = pd.DataFrame()
    Ca_VGCC_AD = pd.DataFrame()
    Ca_CYTO_AD = pd.DataFrame()
    Ca_IP3_AD = pd.DataFrame()

    #time = pd.read_csv("time.txt", header=None, names=["time"])

    for k in tqdm_notebook(range(5,155,5)):

        rel_vesicles_WT = pd.concat([rel_vesicles_WT, pd.read_csv("{}/WT_DualSensor_RelVes_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)    
        rel_rate_WT = pd.concat([rel_rate_WT, pd.read_csv("{}/WT_DualSensor_RelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        reserve_pool_rate_WT = pd.concat([reserve_pool_rate_WT, pd.read_csv("{}/WT_DualSensor_Rate_ReservePool_{}.txt"
                                          .format(filepath, k), header=None, names=["{} Channels".format(k)])], axis=1)
        docked_pool_rate_WT = pd.concat([docked_pool_rate_WT, pd.read_csv("{}/WT_DualSensor_Rate_DockedPool_{}.txt".
                                        format(filepath, k),header=None, names=["{} Channels".format(k)])], axis=1)
        aysnc_rel_rate_WT = pd.concat([aysnc_rel_rate_WT, pd.read_csv("{}/WT_DualSensor_AsyncRelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        sync_rel_rate_WT = pd.concat([sync_rel_rate_WT, pd.read_csv("{}/WT_DualSensor_SyncRelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        spont_rel_rate_WT = pd.concat([spont_rel_rate_WT, pd.read_csv("{}/WT_DualSensor_SpontRelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)        
        slow_rel_rate_WT = pd.concat([slow_rel_rate_WT, pd.read_csv("{}/WT_DualSensor_SlowRelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        fast_rel_rate_WT = pd.concat([fast_rel_rate_WT, pd.read_csv("{}/WT_DualSensor_FastRelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1) 
        rel_proba_WT = pd.concat([rel_proba_WT, pd.read_csv("{}/WT_DualSensor_RelProba_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        RRV_WT = pd.concat([RRV_WT, pd.read_csv("{}/WT_DualSensor_RRV_{}.txt".format(filepath, k),
                                    header=None, names=["{} Channels".format(k)])], axis=1) 
        Ca_VGCC_WT = pd.concat([Ca_VGCC_WT, pd.read_csv("{}/WT_VGCC_Calcium_Dynamics_{}.txt".format(filepath, k),
                                            header=None, names=["{} Channels".format(k)])], axis=1)     
        Ca_CYTO_WT = pd.concat([Ca_CYTO_WT, pd.read_csv("{}/WT_CYTO_Calcium_Dynamics_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(filepath, k)])], axis=1)      
        Ca_IP3_WT = pd.concat([Ca_IP3_WT, pd.read_csv("{}/WT_IP3_Calcium_Dynamics_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1) 



        rel_vesicles_AD = pd.concat([rel_vesicles_AD, pd.read_csv("{}/AD_DualSensor_RelVes_{}.txt".format(filepath, k), 
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        rel_rate_AD = pd.concat([rel_rate_AD, pd.read_csv("{}/AD_DualSensor_RelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        reserve_pool_rate_AD = pd.concat([reserve_pool_rate_AD, pd.read_csv("{}/AD_DualSensor_Rate_ReservePool_{}.txt"
                                          .format(filepath, k), header=None, names=["{} Channels".format(k)])], axis=1)
        docked_pool_rate_AD = pd.concat([docked_pool_rate_AD, pd.read_csv("{}/AD_DualSensor_Rate_DockedPool_{}.txt".
                                        format(filepath, k),header=None, names=["{} Channels".format(k)])], axis=1)
        aysnc_rel_rate_AD = pd.concat([aysnc_rel_rate_AD, pd.read_csv("{}/AD_DualSensor_AsyncRelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        sync_rel_rate_AD = pd.concat([sync_rel_rate_AD, pd.read_csv("{}/AD_DualSensor_SyncRelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        spont_rel_rate_AD = pd.concat([spont_rel_rate_AD, pd.read_csv("{}/AD_DualSensor_SpontRelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)        
        slow_rel_rate_AD = pd.concat([slow_rel_rate_AD, pd.read_csv("{}/AD_DualSensor_SlowRelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        fast_rel_rate_AD = pd.concat([fast_rel_rate_AD, pd.read_csv("{}/AD_DualSensor_FastRelRate_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1) 
        rel_proba_AD = pd.concat([rel_proba_AD, pd.read_csv("{}/AD_DualSensor_RelProba_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        RRV_AD = pd.concat([RRV_AD, pd.read_csv("{}/AD_DualSensor_RRV_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        Ca_VGCC_AD = pd.concat([Ca_VGCC_AD, pd.read_csv("{}/AD_VGCC_Calcium_Dynamics_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        Ca_CYTO_AD = pd.concat([Ca_CYTO_AD, pd.read_csv("{}/AD_CYTO_Calcium_Dynamics_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        Ca_IP3_AD = pd.concat([Ca_IP3_AD, pd.read_csv("{}/AD_IP3_Calcium_Dynamics_{}.txt".format(filepath, k),
                                                 header=None, names=["{} Channels".format(k)])], axis=1)
        
    
    
    rel_vesicles_WT.to_csv("{}/WT_VesiclesReleased_{}.csv".format(folder_to_write, coupling_type),
                           header=False, index=False)
    rel_rate_WT.to_csv("{}/WT_ReleaseRate_{}.csv".format(folder_to_write, coupling_type),
                       header=False, index=False)
    reserve_pool_rate_WT.to_csv("{}/WT_ReservedPoolRelRate_{}.csv".format(folder_to_write, coupling_type),
                                header=False, index=False)
    docked_pool_rate_WT.to_csv("{}/WT_DockedPoolRelRate_{}.csv".format(folder_to_write, coupling_type),
                               header=False, index=False)
    aysnc_rel_rate_WT.to_csv("{}/WT_AsyncRelRate_{}.csv".format(folder_to_write, coupling_type),
                             header=False, index=False)
    sync_rel_rate_WT.to_csv("{}/WT_SyncRelRate_{}.csv".format(folder_to_write, coupling_type),
                            header=False, index=False)
    spont_rel_rate_WT.to_csv("{}/WT_SpontRelRate_{}.csv".format(folder_to_write, coupling_type),
                             header=False, index=False)
    slow_rel_rate_WT.to_csv("{}/WT_SlowRelRate_{}.csv".format(folder_to_write, coupling_type),
                            header=False, index=False)
    fast_rel_rate_WT.to_csv("{}/WT_FastRelRate_{}.csv".format(folder_to_write, coupling_type),
                            header=False, index=False)
    rel_proba_WT.to_csv("{}/WT_ReleaseProba_{}.csv".format(folder_to_write, coupling_type),
                        header=False, index=False)
    RRV_WT.to_csv("{}/WT_RRV_{}.csv".format(folder_to_write, coupling_type), header=False,
                  index=False)
    Ca_VGCC_WT.to_csv("{}/WT_Ca_VGCC_{}.csv".format(folder_to_write, coupling_type),
                      header=False, index=False)
    Ca_CYTO_WT.to_csv("{}/WT_Ca_CYTO_{}.csv".format(folder_to_write, coupling_type),
                      header=False, index=False)
    Ca_IP3_WT.to_csv("{}/WT_Ca_IP3_{}.csv".format(folder_to_write, coupling_type),
                     header=False, index=False)

    
    
    rel_vesicles_AD.to_csv("{}/AD_VesiclesReleased_{}.csv".format(folder_to_write, coupling_type),
                           header=False, index=False)
    rel_rate_AD.to_csv("{}/AD_ReleaseRate_{}.csv".format(folder_to_write, coupling_type),
                       header=False, index=False)
    reserve_pool_rate_AD.to_csv("{}/AD_ReservedPoolRelRate_{}.csv".format(folder_to_write, coupling_type),
                                header=False, index=False)
    docked_pool_rate_AD.to_csv("{}/AD_DockedPoolRelRate_{}.csv".format(folder_to_write, coupling_type),
                               header=False, index=False)
    aysnc_rel_rate_AD.to_csv("{}/AD_AsyncRelRate_{}.csv".format(folder_to_write, coupling_type),
                             header=False, index=False)
    sync_rel_rate_AD.to_csv("{}/AD_SyncRelRate_{}.csv".format(folder_to_write, coupling_type),
                            header=False, index=False)
    spont_rel_rate_AD.to_csv("{}/AD_SpontRelRate_{}.csv".format(folder_to_write, coupling_type),
                             header=False, index=False)
    slow_rel_rate_AD.to_csv("{}/AD_SlowRelRate_{}.csv".format(folder_to_write, coupling_type),
                            header=False, index=False)
    fast_rel_rate_AD.to_csv("{}/AD_FastRelRate_{}.csv".format(folder_to_write, coupling_type),
                            header=False, index=False)
    rel_proba_AD.to_csv("{}/AD_ReleaseProba_{}.csv".format(folder_to_write, coupling_type),
                        header=False, index=False)
    RRV_AD.to_csv("{}/AD_RRV_{}.csv".format(folder_to_write, coupling_type), header=False,
                  index=False)
    Ca_VGCC_AD.to_csv("{}/AD_Ca_VGCC_{}.csv".format(folder_to_write, coupling_type),
                      header=False, index=False)
    Ca_CYTO_AD.to_csv("{}/AD_Ca_CYTO_{}.csv".format(folder_to_write, coupling_type),
                      header=False, index=False)
    Ca_IP3_AD.to_csv("{}/AD_Ca_IP3_{}.csv".format(folder_to_write, coupling_type),
                     header=False, index=False)

    return
```

```python
## Create path for saving processed/packaged data if directory does not exits already
## ideally the path should already exist since simulation automatically saves data to the directory

path = "../../data/SingleAP_data/"

if not os.path.exists(f"{path}Preprocessed/"):
    os.mkdir(f"{path}Preprocessed/")
```

```python
PackageData_SingleAP("SameCoupling", f"{path}Same_Coupling", f"{path}Preprocessed/")
PackageData_SingleAP("AD_HigherCoupling", f"{path}Higher_Coupling_AD", f"{path}Preprocessed/")
PackageData_SingleAP("WT_HigherCoupling", f"{path}Higher_Coupling_WT", f"{path}Preprocessed/")
```
